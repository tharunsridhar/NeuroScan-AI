```python
import sys
from pathlib import Path

import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image

# ============================================================
# UPDATED PROJECT PATH
# ============================================================
CODE_DIR = Path(r"/content/drive/MyDrive/Project work/Code")
if str(CODE_DIR) not in sys.path:
    sys.path.insert(0, str(CODE_DIR))

from core import (
    CLASS_NAMES,
    IMG_SIZE,
    quality_metrics,
    adaptive_model_fusion,
    run_segmentation,
    get_gradcam,
    estimate_size,
    analyze_shape,
    mass_effect,
    overlap_metrics,
    reliability_and_risk,
    reliability_gate,
    confidence_cal,
    get_severity_label,
    clinical_decision,
    rano_assessment,
)

# ============================================================
# MODEL PATHS
# ============================================================
classification_model_paths = {
    "EfficientNetV2-S": Path(r"/content/drive/MyDrive/Project work/models/class_Tumor_v2s_clean.keras"),
    "MobileNetV3": Path(r"/content/drive/MyDrive/Project work/models/class_Tumor_mobilenet_v3.keras"),
    "ConvNeXt Tiny": Path(r"/content/drive/MyDrive/Project work/models/class_Tumor_convnext_tiny_tumor.keras"),
}

segmentation_model_path = Path(r"/content/drive/MyDrive/Project work/models/Segmentation_brisc_effunet.keras")

classification_preprocess = {
    "EfficientNetV2-S": tf.keras.applications.efficientnet_v2.preprocess_input,
    "MobileNetV3": tf.keras.applications.mobilenet_v3.preprocess_input,
    "ConvNeXt Tiny": tf.keras.applications.convnext.preprocess_input,
}

# ============================================================
# IMAGE PATHS
# ============================================================
image_paths = [
    Path(r"/content/drive/MyDrive/Project work/Test Data/test_glioma.jpg"),
    Path(r"/content/drive/MyDrive/Project work/Test Data/test_meningioma.jpg"),
    Path(r"/content/drive/MyDrive/Project work/Test Data/test_pituitary.jpg"),
    Path(r"/content/drive/MyDrive/Project work/Test Data/test_no tumor.jpg"),
]

# ============================================================
# LOAD MODELS
# ============================================================
def load_all_models(classification_model_paths, segmentation_model_path):
    classifiers = {}
    for model_name, model_path in classification_model_paths.items():
        classifiers[model_name] = tf.keras.models.load_model(str(model_path), compile=False)

    seg_model = tf.keras.models.load_model(str(segmentation_model_path), compile=False)

    base_model = None
    head_layers = []

    if "EfficientNetV2-S" in classifiers:
        try:
            effnet_model = classifiers["EfficientNetV2-S"]
            base_model = effnet_model.get_layer("efficientnetv2-s")
            hit = False
            for layer in effnet_model.layers:
                if layer.name == "efficientnetv2-s":
                    hit = True
                    continue
                if hit:
                    head_layers.append(layer)
        except Exception:
            base_model = None
            head_layers = []

    return classifiers, seg_model, base_model, head_layers

# ============================================================
# PREPROCESS HELPER
# ============================================================
def preprocess_for_model(image_pil, preprocess_fn):
    arr = np.array(image_pil.convert("RGB").resize((IMG_SIZE, IMG_SIZE)), dtype=np.float32)
    try:
        arr = preprocess_fn(arr)
    except Exception:
        arr = arr / 255.0
    return np.expand_dims(arr, axis=0)

# ============================================================
# FUNCTION 1 : CLASSIFICATION
# ============================================================
def get_classification_outputs(image_path, classifiers, classification_preprocess):
    image_pil = Image.open(image_path).convert("RGB")
    rgb_384 = np.array(image_pil.resize((IMG_SIZE, IMG_SIZE)), dtype=np.uint8)

    raw_probs = {}
    per_model_scores = {}

    for model_name, model in classifiers.items():
        x = preprocess_for_model(image_pil, classification_preprocess[model_name])
        probs = model.predict(x, verbose=0)[0]

        raw_probs[model_name] = probs
        per_model_scores[model_name] = {
            class_name: round(float(score), 4)
            for class_name, score in zip(CLASS_NAMES, probs)
        }

    quality = quality_metrics(rgb_384)
    first_fusion = adaptive_model_fusion(raw_probs, quality["quality_score"])

    return {
        "image_pil": image_pil,
        "rgb_384": rgb_384,
        "raw_probs": raw_probs,
        "per_model_scores": per_model_scores,
        "quality": quality,
        "first_fusion": first_fusion,
    }

# ============================================================
# FUNCTION 2 : SEGMENTATION
# ============================================================
def get_segmentation_output(image_path, seg_model, base_model, head_layers):
    image_pil = Image.open(image_path).convert("RGB")
    rgb_384 = np.array(image_pil.resize((IMG_SIZE, IMG_SIZE)), dtype=np.uint8)

    eff_input = np.array(image_pil.resize((IMG_SIZE, IMG_SIZE)).convert("RGB"), dtype=np.float32)
    try:
        eff_input = tf.keras.applications.efficientnet_v2.preprocess_input(eff_input)
    except Exception:
        eff_input = eff_input / 255.0
    eff_input = np.expand_dims(eff_input, axis=0)

    heatmap = get_gradcam(eff_input, base_model, head_layers)
    heatmap_resized = cv2.resize(np.asarray(heatmap, dtype=np.float32), (IMG_SIZE, IMG_SIZE))

    seg_mask = run_segmentation(seg_model, str(image_path))

    mask_uint8 = (seg_mask * 255).astype(np.uint8)
    heatmap_uint8 = np.uint8(np.clip(heatmap_resized, 0, 1) * 255)

    seg_colored = cv2.applyColorMap(mask_uint8, cv2.COLORMAP_HOT)
    seg_overlay_bgr = cv2.addWeighted(cv2.cvtColor(rgb_384, cv2.COLOR_RGB2BGR), 0.7, seg_colored, 0.3, 0)
    seg_overlay = cv2.cvtColor(seg_overlay_bgr, cv2.COLOR_BGR2RGB)

    heatmap_colored = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heat_overlay_bgr = cv2.addWeighted(cv2.cvtColor(rgb_384, cv2.COLOR_RGB2BGR), 0.6, heatmap_colored, 0.4, 0)
    heat_overlay = cv2.cvtColor(heat_overlay_bgr, cv2.COLOR_BGR2RGB)

    return {
        "segmentation_mask": seg_mask,
        "heatmap": heatmap_resized,
        "mask_shape": seg_mask.shape,
        "mask_pixel_sum": int(np.sum(seg_mask)),
        "mask_display": mask_uint8,
        "heatmap_display": heatmap_uint8,
        "seg_overlay": seg_overlay,
        "heat_overlay": heat_overlay,
    }

# ============================================================
# FUNCTION 3 : ALL FEATURES
# ============================================================
def get_all_feature_outputs(classification_result, segmentation_result):
    raw_probs = classification_result["raw_probs"]
    quality = classification_result["quality"]

    seg_mask = segmentation_result["segmentation_mask"]
    heatmap_resized = segmentation_result["heatmap"]

    size_info = estimate_size(seg_mask)
    shape_info = analyze_shape(seg_mask)
    mass_info = mass_effect(seg_mask)
    overlap = overlap_metrics(heatmap_resized, seg_mask)

    lesion_context = {
        "area_cm2": size_info["area_cm2"],
        "diameter_cm": size_info["diameter_cm"],
        "irregularity": 0.0 if shape_info is None else shape_info["irregularity"],
        "overlap_score": overlap["overlap_score"],
    }

    final_fusion = adaptive_model_fusion(
        raw_probs,
        quality["quality_score"],
        lesion_context=lesion_context,
    )

    risk_info = reliability_and_risk(
        final_fusion["final_class"],
        final_fusion["fused_confidence"],
        final_fusion["agreement_score"],
        quality["quality_score"],
        size_info,
        shape_info,
        mass_info,
        overlap["overlap_score"],
    )

    gate_info = reliability_gate(quality, final_fusion, overlap, risk_info)

    severity = get_severity_label(
        final_fusion["final_class"],
        final_fusion["fused_confidence"],
        size_info["tumor_percent"],
    )

    confidence_info = confidence_cal(final_fusion["fused_confidence"])

    clinical_info = clinical_decision(
        final_fusion["final_class"],
        severity,
        mass_info,
    )

    rano_info = rano_assessment(
        final_fusion["final_class"],
        size_info,
        shape_info,
    )

    return {
        "quality": quality,
        "size_info": size_info,
        "shape_info": shape_info,
        "mass_effect": mass_info,
        "overlap": overlap,
        "final_fusion": final_fusion,
        "risk_info": risk_info,
        "gate_info": gate_info,
        "severity": severity,
        "confidence_calibration": confidence_info,
        "clinical_decision": clinical_info,
        "rano_assessment": rano_info,
    }

# ============================================================
# DISPLAY FUNCTION
# ============================================================
def display_complete_result(image_path, classification_result, segmentation_result, feature_result):
    print("\n" + "=" * 120)
    print(f"IMAGE ADDRESS : {image_path}")
    print("=" * 120)

    print("\n1. CLASSIFICATION")
    for model_name, scores in classification_result["per_model_scores"].items():
        majority_class = max(scores, key=scores.get)
        majority_score = scores[majority_class]
        print(f"{model_name:<20} -> {majority_class:<12} : {majority_score:.4f}")

    print(f"\nFinal Prediction      : {feature_result['final_fusion']['final_class']}")
    print(f"Fused Confidence      : {feature_result['final_fusion']['fused_confidence']:.4f}")
    print(f"Agreement Score       : {feature_result['final_fusion']['agreement_score']:.4f}")

    print("\n2. SEGMENTATION")
    print(f"Mask Shape            : {segmentation_result['mask_shape']}")
    print(f"Mask Pixel Sum        : {segmentation_result['mask_pixel_sum']}")

    fig, axes = plt.subplots(1, 5, figsize=(24, 5))

    axes[0].imshow(classification_result["rgb_384"])
    axes[0].set_title("Input Image")
    axes[0].axis("off")

    axes[1].imshow(segmentation_result["mask_display"], cmap="gray")
    axes[1].set_title("Segmentation Mask")
    axes[1].axis("off")

    axes[2].imshow(segmentation_result["seg_overlay"])
    axes[2].set_title("Segmentation Overlay")
    axes[2].axis("off")

    axes[3].imshow(segmentation_result["heatmap_display"], cmap="jet")
    axes[3].set_title("Grad-CAM Heatmap")
    axes[3].axis("off")

    axes[4].imshow(segmentation_result["heat_overlay"])
    axes[4].set_title("Grad-CAM Overlay")
    axes[4].axis("off")

    plt.tight_layout()
    plt.show()

    print("\n3. FEATURE OUTPUT")
    print(f"Quality Score         : {feature_result['quality']['quality_score']:.4f}")
    print(f"Tumor Area (cm^2)     : {feature_result['size_info']['area_cm2']}")
    print(f"Diameter (cm)         : {feature_result['size_info']['diameter_cm']}")
    print(f"Tumor Percent         : {feature_result['size_info']['tumor_percent']}")
    print(f"Irregularity          : {None if feature_result['shape_info'] is None else feature_result['shape_info']['irregularity']}")
    print(f"Midline Shift (mm)    : {feature_result['mass_effect']['shift_mm']}")
    print(f"Overlap Score         : {feature_result['overlap']['overlap_score']:.4f}")
    print(f"Severity              : {feature_result['severity']}")
    print(f"Risk                  : {feature_result['risk_info']['risk']}")
    print(f"Reliability Score     : {feature_result['risk_info']['reliability_score']:.4f}")
    print(f"Gate Decision         : {feature_result['gate_info']['acceptance_status']}")

# ============================================================
# ALL-IN-ONE FUNCTION
# ============================================================
def run_complete_test_for_image(image_path, classifiers, seg_model, base_model, head_layers, classification_preprocess):
    classification_result = get_classification_outputs(
        image_path=image_path,
        classifiers=classifiers,
        classification_preprocess=classification_preprocess,
    )

    segmentation_result = get_segmentation_output(
        image_path=image_path,
        seg_model=seg_model,
        base_model=base_model,
        head_layers=head_layers,
    )

    feature_result = get_all_feature_outputs(
        classification_result=classification_result,
        segmentation_result=segmentation_result,
    )

    display_complete_result(
        image_path=image_path,
        classification_result=classification_result,
        segmentation_result=segmentation_result,
        feature_result=feature_result,
    )

    return {
        "image_path": str(image_path),
        "classification_result": classification_result,
        "segmentation_result": segmentation_result,
        "feature_result": feature_result,
    }

# ============================================================
# LOAD MODELS ONCE
# ============================================================
classifiers, seg_model, base_model, head_layers = load_all_models(
    classification_model_paths=classification_model_paths,
    segmentation_model_path=segmentation_model_path,
)

# ============================================================
# RUN FOR ALL IMAGES
# ============================================================
all_results = []

for image_path in image_paths:
    result = run_complete_test_for_image(
        image_path=image_path,
        classifiers=classifiers,
        seg_model=seg_model,
        base_model=base_model,
        head_layers=head_layers,
        classification_preprocess=classification_preprocess,
    )
    all_results.append(result)
```
